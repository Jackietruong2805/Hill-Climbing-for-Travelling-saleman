```python
import random
```

# **Tạo ra giải pháp đường đi ngẫu nhiên**

```python
def randomSolution(map):
  cities = list(range(len(map)))
  solution = []
  for i in range(len(map)):
    randomCity = cities[random.randint(0, len(cities) - 1)]
    solution.append(randomCity)
    cities.remove(randomCity)
  return solution
```

# **Tính độ dài của đường đi**

```python
def routeLength(map, solution):
  routeLength = 0
  for i in range(len(solution)):
    routeLength += map[solution[i-1]][solution[i]]
  return routeLength
```

# **Tìm kiếm các giải pháp lân cận**

```python
def getNeighbors(solution):
  neighbors = []
  for i in range(len(solution)):
    for j in range(i+1, len(solution)):
      neighbor = solution.copy()
      neighbor[i], neighbor[j] = solution[j], solution[i]
      neighbors.append(neighbor)
  return neighbors
```

# **Tìm giải pháp lân cận tốt nhất**

```python
def getBestNeighbor(map, neighbors):
  bestRouteLength = routeLength(map, neighbors[0])
  bestNeighbor = neighbors[0]
  for neighbor in neighbors:
    currentRouteLength = routeLength(map, neighbor)
    if currentRouteLength < bestRouteLength:
      bestRouteLength = currentRouteLength
      bestNeighbor = neighbor
  return bestNeighbor, bestRouteLength
```

# **Giải thuật Hill Climbing cho bài toán người du lịch**

```python
def hillclimbing(map):
  currentSolution = randomSolution(map)
  currentRouteLength = routeLength(map, currentSolution)
  initial = [currentSolution, currentRouteLength]
  neighbors = getNeighbors(currentSolution)
  bestNeighbor, bestNeighborRouteLength = getBestNeighbor(map, neighbors)

  while bestNeighborRouteLength < currentRouteLength:
    currentSolution = bestNeighbor
    currentRouteLength = bestNeighborRouteLength
    neighbors = getNeighbors(currentSolution)
    bestNeighbor, bestNeighborRouteLength = getBestNeighbor(map, neighbors)
  
  return currentSolution, currentRouteLength, initial
```

```python
def main():
  map = [
      [0, 400, 500, 300],
      [400, 0, 300, 500],
      [500, 300, 0, 400],
      [300, 500, 400, 0]
  ]
  currentSolution, currentRouteLength, initial = hillclimbing(map)
  print("Initial Path", initial[0])
  print("Initial Distance", initial[1])
  print("Optimized Path: ", currentSolution)
  print("Optimized Distance: ", currentRouteLength)
if __name__ == "__main__":
  main()
```
